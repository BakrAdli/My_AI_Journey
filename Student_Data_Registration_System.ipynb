```python
#Today I learned the variables and data types and I will get in for a small example
#Student Data Registration System
student_name = input("the student name is : ")                       #string
age = int(input("the student age is :" ))                            #integer
gpa = float(input("the student gpa is :" ))                          #float
while True :                                                         #This loop ensures that the answer is strictly "Yes" or "No"
    status_input = input("Is the student registered at the university or not? ").strip().lower()
    if status_input == "yes":
        is_onrolled = True
        break
    elif status_input == "no":
        is_onrolled = False
        break
    else:
        print("pleas answer with yes or no")
print("\n   Registered student data")
print(f"the name : {student_name}")
print(f"the age : {age}")
print(f"the gpa : {gpa}")
print(f"Registration Status : {is_onrolled}")
with open("student_records.txt","a") as f:                          #The data is written to the file here
    f.write ( f"the student {student_name}, the age {age}, the gpa {gpa},their registration status {is_onrolled} \n")
    print("The student's data has been saved successfully!!")
```
