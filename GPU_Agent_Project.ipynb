```python
#--HELLO WORLD . my name is BAKR ADLI and i summarize everything I learn into templates to document the entire learning journey.
def analyze_gpu(vram):
    if vram >= 16:
        return 'Action: START LOCAL TRAINING'
    elif vram >= 8:
        return 'Action: RUN LOCAL INFERNCE'
    else :
        return 'Action: OFF-LOAD TO CLOUD'

def run_agent():
    try :
        with open ('gpu_list.txt','r') as f:
            lines = f.readlines()

        result = []
        for line in lines :
            vram = int (line.strip())
            action = analyze_gpu (vram)
            result.append(f'GPU {vram}GB -> {action}')

        with open ('ai_report.txt','w') as f:
            for item in result:
                f.write( item + '\n')
        print ('The mission was successfully completed you can see the result in ai_report.txt')
    except FileNotFoundError:
        print ('ERROR  [gpu_list.txt] it does not exist. Please create it first')
if __name__ == '__main__':
    run_agent()
```
