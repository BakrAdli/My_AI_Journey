```python
 # Today,We will update the code and make it more professional by using specialized libraries

# Start with import the libraries
import os
# Create a CSV file to save data
def save_to_file(name,speed):
    with open("processors_data.csv", "a") as f:
        f.write(f"{name},{speed}\n")
# A function to receive data from the user, including error handling
def get_processor_data():
    while True:
        name = input("Enter processor name (or type 'finish' to exit) : ").strip()
        if name.lower() == "finish":
            return None, None
        try:
            speed_val = float(input(f"Enter the speed for{name} (in GHz) : "))
            return name, speed_val
        except ValueError:
            print("Error: speed must be a valid number. please try again.")
def run_database():
    processors = []
    speeds = []
    print("=== Professional Processor Database system ===")
    while True:
        name, speed_val = get_processor_data()
        if name is None:
            break
        processors.append(name)
        speeds.append(speed_val)
        save_to_file(name, speed_val)
        print(f"-->{name} saved successfully !\n")
        if processors:
            print("\n--- Analytics Report ---")
            max_speed = max(speeds)
            idx = speeds.index(max_speed)
    print(f"Fastest: {processors[idx]} at {max_speed} GHz")
    print(f"Total entries: {len(processors)} You can see them in the file processors_data.csv ")

if __name__ == "__main__":
    run_database()
```
