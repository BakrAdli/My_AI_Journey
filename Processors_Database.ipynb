```python
#Update 3.0 reviews the database and adjusts the processor speeds within the folder
import os

def load_from_file():
    processors = []
    speeds = []
    if os.path.exists("processors_data.csv"):
        with open("processors_data.csv", "r") as f:
            for line in f:
                if line.strip():
                    name, speed = line.strip().split(",")
                    processors.append(name)
                    speeds.append(float(speed))
    return processors, speeds

def save_all_to_file(processors, speeds):
    with open("processors_data.csv", "w") as f:
        for i in range(len(processors)):
            f.write(f"{processors[i]},{speeds[i]}\n")

def get_processor_data():
    while True:
        name = input("Enter processor name (or type 'finish' to exit) : ").strip()
        if name.lower() == "finish":
            return None, None
        try:
            speed_val = float(input(f"Enter the speed for {name} (in GHz) : "))
            return name, speed_val
        except ValueError:
            print("Error: speed must be a valid number. please try again.")

def run_database():
    processors, speeds = load_from_file()

    print("=== Professional Processor Database system ===")
    print(f"--> Loaded {len(processors)} processors from previous sessions.\n")

    while True:
        name, speed_val = get_processor_data()

        if name is None:
            break

        if name in processors:
            idx = processors.index(name)
            speeds[idx] = speed_val
            print(f"--> {name} already exists. Speed updated to {speed_val} GHz!\n")
        else:
            processors.append(name)
            speeds.append(speed_val)
            print(f"--> {name} added successfully !\n")

        save_all_to_file(processors, speeds)

    if processors:
        print("\n--- Analytics Report ---")
        max_speed = max(speeds)
        idx = speeds.index(max_speed)
        print(f"Fastest: {processors[idx]} at {max_speed} GHz")
        print(f"Total entries: {len(processors)}")

if __name__ == "__main__":
    run_database()
```
