```python
# Today,we will use lists to create a registration system for processors
# Start with creating empty lists to store data
processors = []
speed = []
print("welcome to the Processors Database")
print("-" * 40)
# Open a loop to continuously receive data
while True:
    proc_name = input("Enter processor name (or type 'finish' to exit) : ").strip()
    if proc_name.lower() == "finish":
        break
# (float) Receiving the processor speed and converting it to a decimal number
    proc_speed = float(input(f"Enter the speed for {proc_name} (in GHz)"))
# Adding the data entered by the user to the lists
    processors.append(proc_name)
    speed.append(proc_speed)
    print(f"-->{proc_name} has been added successfully! \n")
# Final printout after exiting the loop
print("-" * 40)
print(f"Done! We have successfully registered {len(processors)} processors in the Database")
```
